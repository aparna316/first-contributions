```
k=3
colors=["green","red","blue"]
clusters={}

for i in range(k):
  center=12*(2*np.random.random((2,))-1)
  points=[]
  cluster={
      'center':center,
      'points':[],
      'color':colors[i]
  }
  clusters[i]=cluster
```

```
X,y = make_blobs(n_samples=500,centers=3,random_state=42)
plt.scatter(X[:,0],X[:,1])

for i in range(k):
  plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1],
              color=clusters[i]['color'],marker='+')
  plt.show()
```

```
def dist(p1,p2):
  return np.sum((p1-p2)**2)**.5

def assignCustomers(clusters):
  m=X.shape[0]
  for ix in range(m):
    current_dists=[]
    for kx in range(k):
      d=dist(X[ix],clusters[kx]['center'])
      current_dists.append(d)
  current_cluster=np.argmin(current_dists)
  clusters[current_cluster]['points'].append(X[ix])


def updateCenters(clusters):
  for i in range(k):
    points=np.array(clusters[i]['points'])
    if(points.shape[0]>0):
      clusters[i]['center']=np.mean(points,axis=0)

      clusters[i]['points']=[]

  return
```

```
def visualise(clusters):
  for i in range(k):
    plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1],
                    color=clusters[i]['color'],marker='+')
    if len(clusters[i]['points'])>0:
      points=np.array(clusters[i]['points'])
      plt.scatter(points[:,0],points[:,1],color=clusters[i]['color'])

  plt.show()
```

```
assignCustomers(clusters)
visualise(clusters)
updateCenters(clusters)
```
